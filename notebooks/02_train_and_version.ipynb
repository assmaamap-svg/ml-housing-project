```python
# Cellule 1 - Initialiser les chemins du projet
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parent
sys.path.insert(0, str(PROJECT_ROOT / "src"))

import ml_housing
print(f"Package ml_housing importé avec succès. Racine : {PROJECT_ROOT}")
```

```python
# Cellule 2 - Charger les données
from ml_housing.data import load_housing_data
from ml_housing.features import split_features_target, split_train_test

df = load_housing_data()
X, y = split_features_target(df)
X_train, X_test, y_train, y_test = split_train_test(X, y)
df.head()
```

```python
# Cellule 3 - Entraîner le modèle v1
from ml_housing.train import train_model

model = train_model(X_train, y_train)
print("Modèle v1 entraîné")
```

```python
# Cellule 4 - Evaluer le modèle V1 

from ml_housing.evaluate import evaluate_model 

metrics = evaluate_model(model, X_test, y_test) 

metrics 

```

```python
# Cellule 5 - Sauvegarder le modèle V1 et ses métriques 
import json 
import joblib 
import re 

models_dir = PROJECT_ROOT / "artifacts" / "models" 
metrics_dir = PROJECT_ROOT / "artifacts" / "metrics" 
models_dir.mkdir(parents=True, exist_ok=True) 
metrics_dir.mkdir(parents=True, exist_ok=True) 
# Trouver le prochain numéro de version 

existing_versions = [int(re.search(r'v(\d+)', f.name).group(1))  

                for f in models_dir.glob("model_v*.joblib")  

                if re.search(r'v(\d+)', f.name)] 

next_v = max(existing_versions, default=0) + 1 

 

joblib.dump(model, models_dir / f"model_v{next_v}.joblib") 

with open(metrics_dir / f"metrics_v{next_v}.json", "w", encoding="utf-8") as file: 

    json.dump(metrics, file, indent=2) 

print(f"Modèle v{next_v} sauvegardé") 
```
